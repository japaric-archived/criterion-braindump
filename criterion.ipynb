# Glossary

First, some definitions to make sure we are on the same page.

**Population**

What we want to analyze, e.g. the execution time of a function `f()`

**Sample**

A collection of individuals randomly chosen from the population, e.g. 100 measurements of the execution time of `f()`

**Estimators**

A function that, when applied to our sample, will give us an estimate of some parameter of the population. Examples:

* The mean of the sample $(\bar{x})$ is an estimator of the mean of the population ($\mu$)
* The standard deviation of the sample ($s$) is an estimator of the standard deviation of the population ($\sigma$)

# Estimating the cost of reading the clock

Let me first explain why this is important. This is how we time some event:

```
// (I)
let start = precise_time_ns();
for _ in range(0, iters) {
    // black_box is necessary, otherwise the compiler may optimize event() away
    black_box(event());
}
let elapsed = precise_time_ns() - start;
```

In the above snippet, the `elapsed` variable holds the time taken to execute  the `for` loop **plus** the time taken to execute `precise_time_ns()` **once**. (If you can't see why this is the case, you can try executing the snippet `(I)` without the `for` loop, and check that `elapsed` is never zero)

To get the execution time of the event (`event_time`), we'll use the following expression:

```
let event_time = (elapsed - clock_cost) as f64 / iters as f64;  // (II)
```

Where `clock_cost` is the cost of executing `precise_time_ns()`. Therefore, is necessary to know `clock_cost` to find out what `event_time` is.

To measure the `clock_cost`, we'll use snippet `(I)`, but replacing `event()` with `precise_time_ns()`.

```
// (III)
let start = precise_time_ns();
for _ in range(0, iters) {
    black_box(precise_time_ns());
}
let elapsed = precise_time_ns() - start;
```

Now we can compute `clock_cost` from `elapsed`, using the following expression (Obtained by replacing `time_event` with `clock_cost` in expression `(II)`)

```
let clock_cost = elapsed as f64 / (iters + 1) as f64;  // (IV)
```

We need to select `iters` to execute snippet `(III)`. As you may guess the higher the value of `iters` is, the better estimate of `clock_cost` we'll get. Executing the snippet will yield one measurement of `clock_cost`, we can take several measurement to further improve the estimate.

We'll pick a sample of 100 measurements, and `iters` will be automatically selected so that sampling (taking one sample) will take between one and two seconds.

Let's see what this sample looks like.

````python
import json

def load_data(filename):
    with open(filename) as f:
        return array(json.loads(f.read())['data'])

sample = load_data('clock.json')
plot(sample, arange(len(sample)), '.')
xlabel('Time (ns)')
ylabel('N-th data point')
title('clock_cost sample')
````

Let's use this sample to estimate the [Probability Density Function](http://en.wikipedia.org/wiki/Probability_density_function) (a.k.a. PDF, not that [PDF](http://en.wikipedia.org/wiki/Pdf)!) of `clock_cost` using a [Kernel Density Estimator](http://en.wikipedia.org/wiki/Kernel_density_estimation) (a.k.a. KDE, not that [KDE](http://en.wikipedia.org/wiki/Kde)!). We'll need a kernel $K(x)$ and a bandwidth factor $h$ to do this. Let's use the [standard normal distribution](http://en.wikipedia.org/wiki/Standard_normal#Definition) as the kernel (because everybody loves the normal distribution!), and use the [Silverman's rule of thumb](http://en.wikipedia.org/wiki/Kernel_density_estimation#Practical_estimation_of_the_bandwidth) to estimate the bandwidth parameter.

````python
import scipy.stats as stats
def plot_pdf(sample, size):
    k = stats.norm.pdf  # kernel = standard normal distribution
    std_dev = std(sample)  # standard deviation
    n = len(sample)
    h = pow(4.0 / 3.0 / n, 1.0 / 5.0) * std_dev  # bandwidth factor
    q1, q3 = percentile(sample, 25), percentile(sample, 75)  # quartiles
    iqr = q3 - q1  # interquartile sample
    if size == 'small':
        min_x = q1 - 3 * iqr
        max_x = q3 + 3 * iqr
    elif size == 'medium':
        min_x = q1 - 4.5 * iqr
        max_x = q3 + 4.5 * iqr
    elif size == 'big':
        min_x = min(sample)
        max_x = max(sample)
    x = linspace(min_x, max_x, 1000)  # X axis (time)
    pdf = 0  # probability density function
    for x_i in sample:
        pdf += k((x - x_i) / h) / n / h
    plot(x, pdf)
plot_pdf(sample, 'big')
mu = mean(sample)
vbar = ylim()
plot([mu] * 2, vbar)
mu_half = median(sample)
plot([mu_half] * 2, vbar)
legend(['PDF', 'mean', 'median'])
xlabel('time (ns)')
ylabel('likelihood')
title('estimated PDF of clock_cost')
````

Ideally we would like to use the most likely value (peak) of the PDF as `clock_cost`, but computing the PDF is rather expensive, so we'll compromise and use the **median** of the sample instead.

# Outlier classification

There may be outliers in the sample, so we'll perform a [box plot](http://en.wikipedia.org/wiki/Box_plot) analysis to check for their presence. The are many criteria to do a box plot, we'll use the following one:

* $Q_1$ is the first quartile
* $Q_3$ is the third quartile
* $IQR = Q_3 - Q_1$ is the interquartile
* the data points that lie in the range $[Q_1 - 1.5 IQR, Q_3 + 1.5 IQR]$ are considered "normal", and anything that lies outside it is considered an outlier
* we'll use another range $[Q_1 - 3 IQ0R, Q_3 + 3 IQR]$, and classify the **outliers** inside this range as **mild**, and the outliers outside it will be considered **severe**.

````python
Q1, Q3 = percentile(sample, 25), percentile(sample, 75)
IQR = Q3 - Q1
lom, him = Q1 - 1.5 * IQR, Q3 + 1.5 * IQR  # threshold of mild outliers
los, his = Q1 - 3 * IQR, Q3 + 3 * IQR  # threshold of severe outliers
plot(sample, arange(len(sample)), '.')
xlabel('Time (ns)')
plot([lom] * 2 + [him] * 2, list(ylim()) + list(reversed(ylim())))
plot([los] * 2 + [his] * 2, list(ylim()) + list(reversed(ylim())))
legend(['Sample', 'Mild', 'Severe'])
title("fib_10: sample + box plot")
````

# Outlier removal

"To drop, or not drop, that is the question" - The statistician

We'll drop the **severe** outliers from the sample, from what I have seen so far, this usually improves the estimate. (Note: The original criterion library **doesn't** do this)

````python
def without_outliers(sample):
    Q1, Q3 = percentile(sample, 25), percentile(sample, 75)
    IQR = Q3 - Q1
    los, his = Q1 - 3 * IQR, Q3 + 3 * IQR  # threshold of severe outliers
    return array(filter(lambda x: x > los and x < his, sample))
sample = without_outliers(sample)
plot(sample, arange(len(sample)), '.')
xlabel('Time (ns)')
ylabel('N-th data point')
title('fib_10: sample without outliers')
````

Now let's estimate the PDF using the reduced sample.

````python
plot_pdf(sample, 'small')
mu = mean(sample)
vbar = ylim()
plot([mu] * 2, vbar)
mu_half = median(sample)
plot([mu_half] * 2, vbar)
legend(['PDF', 'mean', 'median'])
xlabel('time (ns)')
title('fib_10: estimated PDF')
````

# Bootstrapping

We could report the mean of one sample as a point estimate of the mean of the population, but if we were to take another sample and compute its mean, that value will be different, so should we report the former or the latter?. Instead, it makes more sense to report a [confidence interval](http://en.wikipedia.org/wiki/Confidence_interval) of the mean of the population.

We could use one of these two approaches:
* If the measurements follow a normal distribution (does the previous PDF plot look like a normal distribution?), then we could find the confidence interval of the mean using an analytical expression.
* If that's not the case, then we could resample the population several times, compute the mean on each sample, and then extract the confidence interval from the distribution of sample means using percentiles.

There's no guarantee that the measurements will follow a normal distribution in all the scenarios, so approach 1 is discarded.

Approach 2 can be used for parametric and non-parametric distributions, so we can use it in all the scenarios. However, approach 2 is very expensive because, for example, taking 1000 samples would take around 25 minutes. Luckily, there is a workaround for this.

Instead of taking a new sample from the population, we could treat the sample we already have **as** the population (after all, this sample came from the population and should be a good representation of it), and take new samples from it, this is known as resampling. We'll use a type of resampling known as [bootstraping](http://en.wikipedia.org/wiki/Bootstrapping_(statistics), that consists in sampling the original sample **with** replacement. This means that the resample may contain one data point from the original sample several times, and/or may not include some data point found in the original sample.

Let's resample our original sample 100,000 times (yup, resampling is cheap, plus, the more the merrier) and plot the bootstrapped means (i.e. the mean of each resample).

````python
with open('b-mean.json') as f:
     sample = array(json.loads(f.read()))
plot(sample, arange(len(sample)), '.')
title('Bootstrapped means')
xlabel('Time (ns)')
ylabel('N-th bootstrapped mean')
````

# Confidence interval

According to the [central limit theorem](http://en.wikipedia.org/wiki/Central_limit_theorem) (a.k.a. CLT), if we take several samples from a population and apply an estimator to each sample, the estimated values should follow a normal distribution (if the numbers of samples is high enough). This should also apply to a bootsrapped distribution.

From the bootstrapped distribution, we can extract the confidence interval using percentiles and also the [standard error](http://en.wikipedia.org/wiki/Standard_error) of the estimate (in this example, of the mean)

Let's check if the CLT holds for the boostrapped means (the PDF should look like a normal distribution) and mark the confidence interval.

````python
plot_pdf(sample, 'big')
ci = 0.95  # 95% confidence interval
standard_error = std(sample)
lower_bound = percentile(sample, 50.0 * (1.0 - ci))  # lower bound of the confidence interval
upper_bound = percentile(sample, 50.0 * (1.0 + ci))  # upper bound of the confidence interval
vbar = ylim()
plot([lower_bound] * 2, vbar) 
plot([upper_bound] * 2, vbar)
title('PDF of the bootstrapped means')
xlabel('time (ns)')
legend(['PDF', 'lower', 'upper'])
````

Now we can report the point estimate of the mean along with the standard error and the confidence interval.

````python
print('mean: {} ± {} [{} {}] 95% CI'.format(mu, standard_error, lower_bound, upper_bound))
````
